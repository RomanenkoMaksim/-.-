# Описание проекта
Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

# Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

# Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

# Обозначения признаков

- BeginDate – дата начала пользования услугами,
- EndDate – дата окончания пользования услугами,
- Type – тип оплаты: ежемесячный, годовой и тд,
- PaperlessBilling – безналичный расчет,
- PaymentMethod – способ оплаты,
- MonthlyCharges – ежемесячные траты на услуги,
- TotalCharges – всего потрачено денег на услуги
- Dependents – наличие иждивенцев
- Senior Citizen – наличие пенсионного статуса по возрасту
- Partner – наличие супруга(и)
- MultipleLines – наличие возможности ведения параллельных линий во время звонка

```python
def delta_for_data (data_main, data_not_main, check_column):
    delta_list = [value for value in data_not_main[check_column] if value not in list(data_main[check_column])]
    if len(delta_list) > 0:
        print(f'В признаке {check_column} втростепенного датафрейма присутствуют следующие различия с основным:')
        print(delta_list)
    else:
        print(f'Значения признака {check_column} идентичны')
```

```python
def feautures_hist(data, col_list ):
    fig, axes = plt.subplots(nrows=len(col_list), ncols=1, figsize=(10, 10))
    axes = axes.ravel() 
    for col, ax in zip(col_list, axes):
        data_hist = (
            data.groupby([col,'out']).size() / data.groupby(['out']).size()).reset_index().rename({0:'percent'}, axis=1)
        sns.histplot(
            x=col, hue='out', legend = True,
            palette='PuBuGn', bins = 100,
            data=data_hist, ax = ax)
        ax.set_title(f'Зависимость признака\n {col} \n от категории клиента:')
        ax.set_xlabel(f'{dict_df[col][0]}')
        
    fig.tight_layout()
    plt.show()
```

```python
def time_hist(data, column):
    data_yes = data[data['out'] == 'Yes']
    data_no = data[data['out'] == 'No']
    data_yes[column].hist(bins=100, alpha=0.8, 
                          density=True, histtype ='bar', figsize=(1, 5), xrot=10, label=['Ушедшие'])
    data_no[column].hist(bins=100, alpha=0.7, 
                         density=True, histtype ='bar', figsize=(10, 5), xrot=10, label=['Оставшиеся'])
    plt.legend()
    plt.title(f'Зависимость признака\n {column} \n от категории клиента:')
    plt.show()    
```

```python
#Функция получения имени модели
def get_model_name(model):
    model_name = str(model).split("(")[0].lower()
    model_name  = (re.sub(r'[^a-zA-Z]',' ', model_name).split())
    return max(model_name, key=len)
```

##### [Funct] Функция для получения предсказаний на тесте

```python
def get_model_test_metrics(model, test_features, test_target):
    #Предсказания
    model_predict_proba = model.predict_proba(test_features)[:, 1]
    model_predict = model.predict(test_features)
    #Получение метрик модели
    roc_auc = round(roc_auc_score(test_target, model_predict_proba), 4)
    accuracy = round(accuracy_score(test_target, model_predict), 4)
    f1 = round(f1_score(test_target, model_predict), 4)
    precision = round(precision_score(test_target, model_predict), 4)
    recall = round(recall_score(test_target, model_predict), 4)
    
    return [roc_auc, accuracy, f1, precision, recall]  
```

```python
#Функция для кросвалидации всех моделей
def model_intarations(
    data, 
    models_list = models_list,
    target_feature = 'out',
    halving_cv = 1,
    data_score = df_crosvalidation
    
                     ):   
    best_model_list = []
    #Разделение выборок
    y = data[target_feature]
    X = data.drop(columns=[target_feature])       
    x_train, x_test, y_train, y_test = feautures_split(X, y)
    #Обучение моделей
    for index_model, model in tqdm_notebook(enumerate(models_list), desc='Прогресс кроссвалидации', total = len(models_list) ):
        #Получение имени интерируемой модели
        model_name = get_model_name(model)
        #Кодировщик catboost работает лучше, чем сделанный нами
        if model_name == 'catboostclassifier':
            pipeline_cv = Pipeline([(model_name, model)])
            print('catboost')
        else:
            pipeline_cv = Pipeline([
            ('trans', col_transformer), 
            (model_name, model)
                              ])
               
        #Запуск подбора параметров
        #HalvingGridSearchCV
        if halving_cv == 1:
            grid_search = HalvingGridSearchCV(
                estimator=pipeline_cv,
                param_grid=param_gridsearch[model_name],
                cv = cv_gridsearch,
                scoring = scoring_halving_gs,
                verbose = verbose_gridsearch,
                n_jobs = n_jobs_gridsearch,
                aggressive_elimination=True, # или False, если не указываем min_resources
                min_resources=2600, #или "exhaust" для автоопределения объёма начальных выборок
                factor=factor_hgs
                                            ) 
        #GridSearchCV
        else:
            grid_search = GridSearchCV(
                estimator=pipeline_cv,
                param_grid=param_gridsearch[model_name],
                cv = cv_gridsearch,
                scoring = scoring_gridsearch,
                refit=refit_gridsearch,
                verbose = verbose_gridsearch,
                n_jobs = n_jobs_gridsearch,
            )
        #Обучение            
        if model_name == 'catboostclassifier':
            print('catboost----------')
            grid_search.fit(x_train, y_train, catboostclassifier__eval_set = (x_train, y_train))
        else:
            grid_search.fit(x_train, y_train)
        #Получение статистики кроссвалидаций 
        results = grid_search.cv_results_
        #Индекс лучшей модели
        best_index = grid_search.best_index_

        #Дополнительные метрики на кроссвалидации
        #Так, как у HalvingGridsearch отсутствует функция мультиоценки, получим предсказания на лучшей модели
        #HalvingGridSearchCV
        try:
            #Лучший результат модели на кроссвалидации
            cv_best_score = round(results['mean_test_score'].mean(), 4) #roc_auc
            cv_accuracy = cv_f1 = cv_precision = cv_recall = 'hgs'
        #GridSearchCV
        except:
            cv_best_score = round(results['mean_test_roc_auc'].mean(), 4) #roc_auc
            cv_accuracy = round(results['mean_test_accuracy'].mean(), 4)
            cv_f1 = round(results['mean_test_f1'].mean(), 4)
            cv_precision = round(results['mean_test_f1'].mean(), 4)
            cv_recall = round(results['mean_test_f1'].mean(), 4)
        
        
               
        #Получение временных значений
        fit_time = round(results['mean_fit_time'][best_index], 4)
        score_time = round(results['mean_score_time'][best_index], 4)
        
        data_score.loc[len(data_score.index)] =([
            model_name, cv_best_score,              #основные
            cv_accuracy, cv_f1, cv_precision,  cv_recall]+      #дополнительные
            get_model_test_metrics(grid_search, x_test, y_test) +     #тест лучшей модели
            [fit_time, score_time, list(grid_search.best_params_.values())] #лучшие гиперпараметры  
                )
        best_model_list.append(grid_search)
        display(data_score.tail(1), grid_search.best_params_)
        
    return data_score, best_model_list
```
